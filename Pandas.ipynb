```python
df = data.copy()
df.columns = ['sl', 'sw', 'pl', 'pw', 'flower_type']

# describe function doesn't show string ke description
df.describe()
```

```python
# adding a new column with user defined properties
# adds a col diff_pl-pw which has diff of pl and pw values
df['diff_pl-pw'] = df['pl'] - df['pw'] 
df.tail()
```

```python
# HANDLING NAN
df.iloc[2:4, 2:4] = np.nan # marked entries as null using numpy's constant nan
print(df.head())

# pl and pw ka count is 147 because of two NaN each, sl and sw are still 149
df.describe()
```

```python
# removes the ROWS which have null entries in them
df.dropna(inplace = True)
df.head()

df.reset_index(drop = True, inplace = True)
df.head()
```

```python
# filling NA entries
df.iloc[2:4, 2:4] = np.nan
df.head()

# mean of pl and pw columns filled in row 2->3
df.pl.fillna(df.pl.mean(), inplace = True)
df.pw.fillna(df.pw.mean(), inplace = True)
df.head()
```

```python
df.iloc[2:4, 2:4] = np.nan
df.head()

# filling na with mean values from columns whose flower type is iris-setosa
a = df[df.flower_type == 'Iris-setosa']
df.pl.fillna(a.pl.mean(), inplace = True)
df.pw.fillna(a.pw.mean(), inplace = True)
df.head()
```

```python
# string based data
df["Gender"] = "female"
df.head()

# making the first 10 entries of gender column (5th) as male
df.iloc[0:10, 5] = "male"
df.head()
df.tail()

# applying functions to string based columns
def f(s):
    if(s=="male"):
        return 0
    else:
        return 1

# makes a new column which applied changed
df["sex"] = df.Gender.apply(f)
df.tail()

# removing Gender column
del df["Gender"]
df.head()
```
