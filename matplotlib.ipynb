```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3]
y = [2,4,6]

# (scatter) and (scatter and show) statements both show the points
plt.scatter(x, y)
# plt.show()
```

```python
x = [1,2,3]
y = [2,4,6]

# shows only the line connecting the points
plt.plot(x, y)
plt.show()
```

```python
x = [1,2,3,4]
y = [2,4,6,10]

x2 = [0,2,3]
y2 = [2,7,6]

# shows both the points and the line connecting the points
plt.scatter(x, y)

# two lines can be  plotted together
# third arg in plot decides colour and type of line:- first char : colour, second char : type

# b - blue, r - red, y - yellow etc
# o : points, - : normal line, -- : dashed line, ^ : triangle, -. : dash dotted line etc
plt.plot(x, y, "y--")
plt.plot(x2, y2)
```

```python
# plot function x^3
import numpy as np

# gives plot as a combination of lines passing from given 4 points
# x = np.array([1,2,3,4])

# arange function takes start, stop and step : 0, 0.1, 0.2, .. ,4.9
x = np.arange(0, 5, 0.1)
y = x**3
plt.plot(x,y, "yo")
```

```python
x = [3,4,5]

# no y given
# in such cases: x is treated as Y and X becomes [0,1,..,n-1] i.e. here [0,1,2]
plt.plot(x)
```

```python
# customizing graphs
x = np.arange(0, 5, 0.1)
y = x**3
y2 = x**2
# plt.plot(x,y, color = "red", marker="o")
plt.plot(x,y, color = "red", linewidth = 1.5, label = "x^3")

plt.plot(x,y2, color = "green", linewidth = 2.5, label = "x^2")

# adding labels for X axis and Y axis
plt.ylabel("Y")
plt.xlabel("X")

# adding labels to distinguish between different curves
plt.legend()

# restricting range of X and Y axes: X till 10, Y till 100 (stops when any one of them goes out of bound)
# plt.axis([0,10,0,100])

# makes a grid
plt.grid()

# adding text on a pair of coordinates on the plot
plt.text(2,80, "anj", fontsize = 15)

plt.title("matplotlib demo")
plt.show()
```

```python
# pie charts

# labels for info
labels = ["ashwamedh", "mirage", "alliance", "debsoc"]

# given 4 divisions, plots by taking ratios i.e. 3/16, 4/16, 2/16 and 7/16
sizes = [3,4,2,7]

# changing colors of divisions
colors = ["orange", "pink", "yellow", "green"]

# the pie chart is a bit flat, not perfectly round
# autopct shows the percentage split within section

# to make segment emerge out with value ranging from 0 to 1
explode = [0.1, 0, 0.2, 0]

# by default graph starts from positive X and moves in anticlock sense
# pass (counterclock = False) in plt.pie() to make it clockwise

# %% to add a percentage sign at the end; .1 means 1 place after decimal
# startangle = 100 makes the graph at 100 from positive X axis
plt.pie(sizes, explode = explode, colors=colors, labels=labels, autopct="%.1f%%", startangle = 100)

plt.title("societies at nsit")

# to make it perfectly round
plt.axis("equal")
plt.show()
```
